```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
from pylab import * 
```

```python
data = pd.read_csv("SentimentTweets.csv")
data.head()
```

```python
favorites = data.iloc[:,1]
```

```python
retweets = data.iloc[:,2]
```

```python
friends = data.iloc[:,4]
followers = data.iloc[:,5]
```

```python
compound = data.iloc[:,7]
pos = data.iloc[:,8]
neu = data.iloc[:,9]
neg = data.iloc[:,10]
```

```python
metadata = [favorites, retweets, friends, followers]
sentiment_scores = [compound, pos, neu, neg]
```

```python
data = {'favorites': favorites,
        'retweets': retweets,
        'friends': friends,
        'followers': followers,
        'compound': compound,
        'pos': pos,
        'neu': neu,
        'neg': neg
        }
```

```python
df = pd.DataFrame(data, columns=['favorites', 'retweets', 'friends', 'followers', 'compound', 'pos', 'neu', 'neg'])
```

```python
corrMatrix = df.corr()
print (corrMatrix)
```

```python
sn.heatmap(corrMatrix, annot=True)
plt.show()
```

indicates that the stronger the sentiment, not necessarily mean more people will like or retweet it

```python
metadata_labels = ['favorites', 'retweets', 'friends', 'followers']
sentiment_scores_labels = ['compound', 'pos', 'neu', 'neg']
```

```python
count1 = 0
count2 = 0
for i in metadata:
    count2 = 0
    for j in sentiment_scores:
        title = metadata_labels[count1] + " vs "+ sentiment_scores_labels[count2]
        plt.figure()
        plt.title(title)
        plt.scatter(j,i)
        count2 = count2 + 1
    count1 = count1 + 1
        
```

```python
plt.scatter(compound,favorites)
plt.ylim(0,750)
```

```python
count = 0
for j in sentiment_scores:
    plt.figure()
    title =  sentiment_scores_labels[count] + " Density"
    plt.title(title)
    plt.hist(j)
    count = count + 1
        
```

```python
m,b = polyfit(compound, favorites, 1) 
plot(compound, favorites, 'yo', compound, m*compound+b, '--k') 
ylim(0,1000)
show() 
```
